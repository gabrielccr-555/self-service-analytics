Também é possível refinar a leitura de dados no momento de carregamento do arquivo!

```python
# Remove os espaços
df["id_cliente"] = df["id_cliente"].str.strip()
df["produto"] = df["produto"].str.strip()
df["nome"] = df["nome"].str.strip()
df
```

```python
# Tratar o problema de case sensitive (maiúsculas / minúsculas)
df["nome"] = df["nome"].str.title()
df["produto"] = df["produto"].str.upper()
df["sistema_origem"] = df["sistema_origem"].str.lower()
df
```

```python
# Remover linhas com valor não definidos;
# Recriando o df eliminando registros onde a coluna id_cliente tem valores NaN
df = df.dropna(subset="id_cliente")
df
```

```python
df["sistema_origem"] = df["sistema_origem"].fillna("outro")
df
```

```python
df["quantidade"][0] = "-"
df
```

```python
# Iremos trabalhar com a quantidade agora!
# info diz que quantidade é object, vamos converter para número
df.info()
```

```python
df["quantidade"] = pd.to_numeric(df["quantidade"], errors="coerce")
df
```

```python
# Vendo info novamente, percebemos que a conversão foi feita!
df.info()
```

Função para formatar um valor de data aplicando mais de uma máscara de formatação:

```python
def parse_data(valor):
  formatos = ["%Y-%m-%d", "%d/%m/%Y"]
  for formato in formatos:
    try:
      return pd.to_datetime(valor, format=formato)
    except:
      pass
    return pd.NaT
```

Aplicar a função **parse_data**

```python
df['data'] = df['data'].apply(parse_data)
df = df.dropna(subset="data")
df
```

Removendo linhas duplicadas (por exemplo linhas 2 e 3)

```python
df = df.drop_duplicates()
df
```

```python
pd.pivot_table(
    df,
    values = 'total',
    index = 'id_cliente',
    columns = 'sistema_origem',
    aggfunc = 'sum',
    fill_value = 0
)
```
